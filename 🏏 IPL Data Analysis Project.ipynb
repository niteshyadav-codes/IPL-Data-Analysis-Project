```python
df[["team1","toss_winner"]]=df[["team1","toss_winner"]].replace({
    "Royal Challengers Bengaluru":"Royal Challengers Bangalore",
    "Rising Pune Supergiants":"Rising Pune Supergiant",
    "Delhi Daredevils":"Delhi Capitals",
    "Kings XI Punjab":"Punjab Kings"

})
df["team2"]=df["team1"].replace({
    "Royal Challengers Bengaluru":"Royal Challengers Bangalore",
    "Rising Pune Supergiants":"Rising Pune Supergiant",
    "Delhi Daredevils":"Delhi Capitals",
    "Kings XI Punjab":"Punjab Kings"

})
df["winner"]=df["winner"].replace({
    "Royal Challengers Bengaluru":"Royal Challengers Bangalore",
    "Rising Pune Supergiants":"Rising Pune Supergiant",
    "Delhi Daredevils":"Delhi Capitals",
    "Kings XI Punjab":"Punjab Kings"

})
```

```python
print("Total team played in IPL since IPL start:")
df["team1"].unique()
```

```python
print("No. of win by each team")
df["winner"].value_counts()
```

```python
winner_counts=df["winner"].value_counts()
print('Maximums matches win by a team')
print(winner_counts.head(1))
print("Mumbai Indians IPL history ki sabse successful teams me se ek hai kyunki isne sabse zyada matches jeete hain.")
```

```python
print("Minimum matches win by a team")
print(winner_counts.tail(1))
print(" Kochi Tuskers Kerala  ne IPL me sabse kam matches jeete hain, jo iski comparatively weak performance ko dikhata hai.")
```

```python
toss_count=df["toss_winner"].value_counts()
print("Maximum Toss win by a team")
print(toss_count.head(1))
print("Mumbai Indians ne sabse zyada toss jeete hain, jo IPL me unki frequent participation aur luck factor ko dikhata hai.")
```

```python
same=df[df["toss_winner"]==df["winner"]]

percentage = (len(same) / len(df)) * 100

print(f" Insight\nIPL me lagbhag {percentage} matches me toss jeetne wali team ne match bhi jeeta. Isse pata chalta hai ki toss jeetna match jeetne ki guarantee nahi hai. Match ka result team ki performance, batting, bowling aur match conditions par bhi kaafi depend karta hai.")
```

```python
print("Most Player of the match win by batsman")
pom_count=df["player_of_match"].value_counts()
print(pom_count.head(1))
print("")
```

###  Key Insights

##### • Most Successful Team: Mumbai Indians won the highest number of matches in IPL history, making them one of the most successful franchises in the tournament.

##### • Most Successful Player: AB de Villiers won the highest number of Player of the Match awards, showcasing his consistent match-winning performances.

##### • Highest Run Scorer: Virat Kohli is the highest run scorer in IPL history, reflecting his remarkable consistency across seasons. 

##### • Highest Wicket Taker: Yuzvendra Chahal has taken the highest number of wickets, making him one of the most successful bowlers in IPL.
